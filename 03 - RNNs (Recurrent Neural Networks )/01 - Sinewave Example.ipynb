## Sine Wave

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(0, 50, 501)
y = np.sin(x)
```

```python
plt.plot(x, y);
plt.title('SineWave');
```

## Dataframe

```python
df = pd.DataFrame(data=y, index=x, columns=['Sine'])
df.head()
```

# Train Test Split
- pembagian data train test split yang berbeda pada umumnya

```python
len(df)
```

```python
# data test 10 persen
test_percent = 0.1
```

```python
len(df)* test_percent
```

```python
# pembulatan data test
test_point = round(len(df) * test_percent)
test_point
```

```python
test_index = len(df) - test_point
test_index
```

```python
train = df.iloc[:test_index]
test = df.iloc[test_index:]
```

```python
train
```

```python
test
```

## Preprocessing data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler_train = scaler.fit_transform(train)
scaler_test = scaler.transform(test)
```

```python
scaler_train
```

```python
scaler_test
```

## Timeseries Generator

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
help(TimeseriesGenerator)
```

```python
# batch pertama
X, y = generator[0]
```

```python
length = 50
generator = TimeseriesGenerator(scaler_train, scaler_train, length= length, batch_size=1)
```

# Create Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, SimpleRNN
```

```python
n_features = 1
```

```python
model = Sequential()

#Simple RNN layers
model.add(SimpleRNN(50, input_shape=(length, n_features)))

# output layers
model.add(Dense(1))

# compile model
model.compile(loss='mse', optimizer='adam')
```

```python
model.summary()
```

## Fitting Model Generator

```python
model.fit(generator, epochs=5, verbose=1)
```

## Evaluation

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot(title='loss');
```

## Add Earlystopping and validation generator


```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=2)
```

```python
# Karena panjannya harus 1 ukuran lebih kecil maka, kita harus mendefinisakn ulang ukuran panjangnya
length = 49

generator = TimeseriesGenerator(scaler_train, scaler_train, length=length, batch_size=1)

validation_generator = TimeseriesGenerator(scaler_test, scaler_test, length=length, batch_size=1)
```

## Model LSTM

```python
model = Sequential()

model.add(LSTM(50, input_shape=(length, n_features)))

model.add(Dense(1))

model.compile(loss='mse', optimizer='adam')
```

```python
model.fit(generator, epochs=20, 
          validation_data = validation_generator,
          callbacks =[early_stop])
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot(title='loss');
```

## Prediction and evaluation

```python
test_prediction =[]
first_evaluation_batch = scaler_train[-length:]
current_batch = first_evaluation_batch.reshape((1, length, n_features))

for i in range(len(test)):
    # curent pred
    current_pred = model.predict(current_batch)[0]
    
    # store prediction
    test_prediction.append(current_pred)
    
    # update bacth include prediction  and drop first values
    current_batch = np.append(current_batch[:, 1:, :], [[current_pred]], axis=1)
```

# Forecasting
Forecast into unknown range. We should first utilize all our data, since we are now forecasting!

```python
full_scaler = MinMaxScaler()
```

```python
scaler_full_data = full_scaler.fit_transform(df)
```

```python
length = 50
generator = TimeseriesGenerator(scaler_full_data,scaler_full_data, length=length, batch_size=1)
```

```python
model = Sequential()

model.add(LSTM(50, input_shape=(length, n_features)))

model.add(Dense(1))

model.compile(loss='mse', optimizer='adam')
```

```python
model.fit(generator, epochs=6)
```

```python
test_prediction =[]
first_eval_batch = scaler_full_data[-length:]
current_batch = first_eval_batch.reshape((1, length, n_features))

for i in range(len(test)):
    # prediction
    current_pred = model.predict(current_batch)[0]
    
    # store predict
    test_prediction.append(current_pred)
    
    # update batch include prediction and drop first values 
    current_batch = np.append(current_batch[:, 1:, :], [[current_pred]], axis=1)
```

```python
true_forecast = scaler.inverse_transform(test_prediction)
```
